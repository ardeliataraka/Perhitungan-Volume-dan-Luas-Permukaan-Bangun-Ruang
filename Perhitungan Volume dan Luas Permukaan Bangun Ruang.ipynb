```python
print ("Program Menghitung Volume dan Luas Permukaan Bangun Ruang")
print ("Chosee one option")
print ("1.) Bangun Ruang Kubus")
print ("2.) Bangun Ruang Balok")
print ("3.) Bangun Ruang Prisma Segitiga")
print ("4.) Bangun Ruang Limas Segiempat")
print ("5.) Bangun Ruang Kerucut")
print ("6.) Bangun Ruang Tabung")
print ("7.) Bangun Ruang Bola")

p=int(input('input your option from 1 to 7'))

if p==1:
    print ('1.) Bangun Ruang Kubus')
    s=int(input('sisi='))
    luas=s**2*6
    volume=s**3
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==2:
    print ('2.) Bangun Ruang Balok')
    p=int(input('panjang='))
    l=int(input('lebar='))
    t=int(input('tinggi='))
    luas=2*(p*l+p*t+l*t)
    volume=p*l*t
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==3:
    print ('3.) Bangun Ruang Prisma Segitiga')
    r1=int(input('alas segitiga ='))
    r2=int(input('tinggi segitiga ='))
    r3=int(input('garis miring segitiga ='))
    T=int(input('tinggi prisma ='))
    luas=(r1*r2)+((r1+r2+r3)*T)
    volume=(1/2*r1*r2)*T
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==4:
    print ('4.) Bangun Ruang Limas Segiempat')
    s=int(input('sisi ='))
    a=int(input('alas ='))
    t=int(input('tinggi ='))
    luas=s**2+(4*(1/2*a*t))
    volume=1/3*s**2*t
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==5:
    print ('5.) Bangun Ruang Kerucut')
    pi=22/7
    r=int(input('jari-jari ='))
    s=int(input('sisi ='))
    t=int(input('tinggi ='))
    luas=pi*r*s+pi*r**2
    volume=1/3*pi*r**2*t
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==6:
    print ('6.) Bangun Ruang Tabung')
    pi=22/7
    r=int(input('jari-jari ='))
    t=int(input('tinggi ='))
    luas=(2*pi*r*t)+(2*pi*r**2)
    volume=pi*r**2*t
    print('Luas permukaan =',luas)
    print('Volume bangun =',volume)

elif p==7:
    print ('7.) Bangun Ruang Bola')
    pi=22/7
    r=int(input('jari-jari ='))
    luas=4*pi*r**2
    volume=4/3*pi**3
    print('Luas permukaan =',luas)
    print('Volume Bangun =',volume)

else:
    print('sorry,we cant calculate it')
```
